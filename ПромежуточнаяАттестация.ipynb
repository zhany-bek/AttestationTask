```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
vals = data['whoAmI'].unique()

for val in vals:
  data[val] = 0

for i, row in data.iterrows():
  data.at[i, row['whoAmI']] = 1

data.drop(columns=['whoAmI'], inplace=True)
data.head()
```
